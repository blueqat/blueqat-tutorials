ここでは、独自のバックエンドを作成する方法を説明します。

カスタムバックエンドは `Backend` を継承したクラスで、単一の `run(self, gates, n_qubits, *args, **kwargs)` メソッドを実装します。このメソッドはゲート操作のリストと量子ビット数を受け取り、好きな結果を返すことができます。今回は、簡単なゲートの説明を返すバックエンドを作ってみます。

```python
from blueqat import Circuit, BlueqatGlobalSetting
from blueqat.backends.backendbase import Backend

class MyBackend(Backend):

    def run(self, gates, n_qubits, *args, **kwargs):
        ctx = ["start"]
        for gate in gates:
            ctx.append(gate.lowername)
        ctx.append("end")
        return "-".join(ctx)

BlueqatGlobalSetting.register_backend('mybackend', MyBackend)
```

バックエンドは簡単に作成できます。早速試してみましょう。

```python
Circuit().h[0].cx[0,1].run(backend="mybackend")
```
