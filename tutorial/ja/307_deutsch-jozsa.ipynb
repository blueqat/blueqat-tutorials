# ドイチ・ジョザのアルゴリズム

ドイチのアルゴリズムを一般化した、ドイチ・ジョザのアルゴリズムについて説明します。
$2^n$ 通りの入力(00...000 から 11...111)を取り得る関数 $f$ について、以下のいずれかが成り立つと仮定します。

1. $f(x)$ がすべての入力で同じ値になる。
つまり、すべての $x$ について $f(x)=0$ であるか、すべての $x$ について $f(x)=1$ である。

2. $f(x)$ が入力の半分で異なる値になる。
つまり、$2^{n-1}$ 個の入力 $x$ について $f(x)=0$ であり、残りについて $f(x)=1$ である。

このアルゴリズムは、オラクルが上記の1.か2.かを判定します。

回路は以下の通りです。

初期状態として、$n$個の量子ビット $\lvert 0\rangle$ と1個の量子ビット $\lvert 1\rangle$ を用意します。

状態を確認してみましょう。

$$
\begin{align}
\lvert \psi_1\rangle &= \biggl(\otimes^n H\lvert 0\rangle \biggr)\otimes H \lvert 1\rangle \\
&= \frac{1}{ \sqrt{2^{n+1}} } \sum^{2^n}_{x\in \{ 0, 1 \}^n} \bigl( \lvert x\rangle \otimes (\lvert 0\rangle - \lvert 1\rangle) \bigr) \\
&= \frac{1}{ \sqrt{2^{n+1}} } \sum^{2^n}_{x\in \{ 0, 1 \}^n} \lvert x\rangle \otimes \lvert 0\rangle - \frac{1}{ \sqrt{2^{n+1}} } \sum^{2^n}_{x\in \{ 0, 1 \}^n} \lvert x\rangle \otimes \lvert 1\rangle \\
\end{align}
$$


$n$番目の量子ビットについて、$\lvert 0\rangle$ と $\lvert 1\rangle$ を入れ替えると全体の符号が変わることが分かります。

次に、$\psi_2$ を考えます。
$f(x)$ は $n$番目の量子ビットに作用するため、

$$
f(x) = 0 \to n\text{番目の量子ビットは反転しない。}
$$
$$
f(x) = 1 \to n\text{番目の量子ビットは反転する。}
$$

各項を $f(x) = 0$ または $1$ で場合分けすると、

$$
\frac{1}{ \sqrt{2^{n+1}} } \sum^{2^n}_{x\in \{ 0, 1 \}^n} \lvert x\rangle \otimes \lvert 0\rangle \xrightarrow{U_f} \frac{1}{ \sqrt{2^{n+1}} } \sum_{f(x)=0} \lvert x\rangle \otimes \lvert 0\rangle + \frac{1}{ \sqrt{2^{n+1}} } \sum_{f(x)=1} \lvert x\rangle \otimes \lvert 1\rangle
$$

$$
\frac{1}{ \sqrt{2^{n+1}} } \sum^{2^n}_{x\in \{ 0, 1 \}^n} \lvert x\rangle \otimes \lvert 1\rangle \xrightarrow{U_f} \frac{1}{ \sqrt{2^{n+1}} } \sum_{f(x)=0} \lvert x\rangle \otimes \lvert 1\rangle + \frac{1}{ \sqrt{2^{n+1}} } \sum_{f(x)=1} \lvert x\rangle \otimes \lvert 0\rangle
$$

$f(x)=1$ のとき $n$番目の量子ビットが反転することが分かります。

したがって、

$$
\lvert \psi_2\rangle = \frac{1}{\sqrt{2^n}} \biggl( \sum_{f(x)=0}\lvert x\rangle - \sum_{f(x)=1}\lvert x\rangle \biggr) \otimes \frac{1}{\sqrt{2}} (\lvert 0\rangle - \lvert 1\rangle)
$$

「1. $f(x)$ がすべての入力で同じ」場合、$f(x)=0$ または $f(x)=1$ のどちらかの項が消えます。
したがって、

$$
\begin{align}
\lvert \psi_2 \rangle &= \pm \frac{1}{ \sqrt{2^{n}} } \sum^{2^n}_{x\in \{ 0, 1 \}^n} \lvert x\rangle \otimes \frac{1}{\sqrt{2}} (\lvert 0\rangle - \lvert 1\rangle) \\
&= \pm \biggl( \otimes^{n} H \lvert 0\rangle \biggr) \otimes H \lvert 1\rangle
\end{align}
$$

結果として、$\lvert \psi_3 \rangle$ は

$$
\lvert \psi_3 \rangle = \pm \lvert 00...00 \rangle \otimes H\lvert 1 \rangle
$$

この場合、0番目から$n$番目までの量子ビットの測定結果はすべて"0"になります。

「2. $f(x)$ が入力の半分で異なる」場合、$f(x)$ によって符号が異なるため、

$$
\lvert \psi_2 \rangle = \frac{1}{ \sqrt{2^{n}} } \sum^{2^n}_{x\in \{ 0, 1 \}^n} (-1)^{f(x)} \lvert x\rangle \otimes \frac{1}{\sqrt{2}} (\lvert 0\rangle - \lvert 1\rangle)
$$

$$
\begin{align}
\lvert \psi_3 \rangle &= \frac{1}{ \sqrt{2^{n}} } \sum^{2^n}_{x\in \{ 0, 1 \}^n} (-1)^{f(x)} (\otimes^nH) \lvert x\rangle \otimes \frac{1}{\sqrt{2}} (\lvert 0\rangle - \lvert 1\rangle) \\
&= \frac{1}{ 2^{n} } \sum^{2^n}_{x\in \{ 0, 1 \}^n} (-1)^{f(x)} \bigl( \sum^{2^n-1}_{y=0}(-1)^{x\cdot y}\lvert y \rangle \bigr) \otimes \frac{1}{\sqrt{2}} (\lvert 0\rangle - \lvert 1\rangle)
\end{align}
$$

ここで、0番目から$n$番目までの測定結果がすべて"0"になる確率 $\rm{Pr}(0)$ を考えます。
これは状態 $\lvert y \rangle\ (y=0)$ の振幅の絶対値の2乗です。

$y=0$ のとき $x\cdot y=0$ であるため、

$$
\mathrm{Pr}(0) = \biggl| \frac{1}{2^n} \sum^{2^n}_{x\in \{ 0, 1 \}^n} (-1)^{f(x)} \biggr|^2
$$

$f(x)$ が入力の半分で異なる場合、すべての項が打ち消し合うため $\mathrm{Pr}(0) =0$ となります。
したがって、0番目から$n$番目までの量子ビットを測定した結果がすべて"0"になることはありません。

以上より、($n-1$)番目までの量子ビットの測定結果がすべて0であるかどうかによって、オラクルを判定できます。

Blueqatでこれを実装してみましょう。

```python
from blueqat import Circuit
import numpy as np
from collections import Counter as _Counter

# Compatibility note: the new blueqat SDK reports measurement bitstrings with
# qubit 0 as the right-most character (standard binary ordering), while this
# notebook (like the old blueqat 0.3.x) reads qubit 0 as the left-most
# character. Patch Circuit.run once so sampled bitstrings keep matching the
# qubit ordering used throughout this notebook's explanation.
if not getattr(Circuit, '_qubit0_leftmost_patched', False):
    _original_run = Circuit.run
    def _run_qubit0_leftmost(self, *args, **kwargs):
        result = _original_run(self, *args, **kwargs)
        if isinstance(result, _Counter):
            return _Counter({key[::-1]: count for key, count in result.items()})
        return result
    Circuit.run = _run_qubit0_leftmost
    Circuit._qubit0_leftmost_patched = True
```

1.と2.それぞれの場合についてオラクルを実装してみましょう。

1.の場合、($n-1$)番目までの量子ビットの値に関わらず、$n$番目の量子ビットを反転させます。

2.の場合、($n-1$)番目までの各量子ビットを制御ビット、$n$番目の量子ビットをターゲットビットとする $CX$ ゲートを $n$個用意します。
この場合、($n-1$)番目までの量子ビットの値の半分について $n$番目の量子ビットが反転し、残りの半分については変化しません。

```python
def oracle_1(c):
    n = c.n_qubits
    c.x[n-1]
    
def oracle_2(c):
    n = c.n_qubits
    for i in range(n-1):
        c.cx[i, n-1]
```

以下がアルゴリズムの本体です。
まず、オラクルが1.と2.のどちらであるかを乱数で決めます。これが求めたい正解となります。

次に、ドイチ・ジョザのアルゴリズムを使ってオラクルを判定します。

最後に、判定結果とあらかじめ乱数で選んだオラクルが一致するか確認します。

```python
n = 4

c = Circuit(n + 1)
c.x[n].h[:]

if np.random.rand() > 0.5:
    oracle_1(c)
    oracle = "f(x) = 1 for all x."
else:
    oracle_2(c)
    oracle = "f(x) = 0 for half x and f(x) = 1 for others."

c.h[:]
res = c.m[:].run(shots = 1000)

print("Oracle:", oracle)
print("Results of quantum circuit:", res)

if [arr[:n] for arr in res.keys()] == ['0'*n] and oracle == "f(x) = 1 for all x.":
    print("OK")
elif [arr[:n] for arr in res.keys()] != ['0'*n] and oracle == "f(x) = 0 for half x and f(x) = 1 for others.":
    print("OK")
else:
    print("incorrect")
```

以上により、ドイチ・ジョザのアルゴリズムを使ってオラクルを判定することができました。
